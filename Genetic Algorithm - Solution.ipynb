```python
def draw_polygon(generation):
    window = Screen()
    window.title('Genetic Algorithm')
    window.screensize(600, 400, bg='black')
    window.colormode(255)
    draw = Turtle()
    draw.hideturtle()
    draw.pencolor('white')
    draw.speed(8)
    def draw_the_shape(optima):
        angles = 360 / optima
        draw.pendown()
        for i in range(optima):
            if optima <= 5:
                draw.forward(25)
            elif optima <= 8:
                draw.forward(20)
            elif optima <= 10:
                draw.forward(14)
            elif optima <= 15:
                draw.forward(10)
            draw.left(angles)
    x, y = 0, 0
    for i in generation:
        x, y = randint(-300, 200), randint(-300, 200)
        r, g, b = randint(0, 255), randint(0, 255), randint(0, 255)
        draw.pencolor(r,g,b)
        draw.penup()
        draw.goto(x,y)
        draw_the_shape(i)
    draw.penup()
    draw.goto(x-10,y-12)
    draw.write("Final Shape")
    window.mainloop()
```

```python
class GeneticAlgorithm:
    
    def __init__(self, size_of_population, optimal_chromosome):
        self.size_of_population = size_of_population
        self.optimal_chromosome = f'{optimal_chromosome:08b}'
        self.population = []
        self.iterations = 0
    
    def fitness_check(self, chromosome):
        fitness = 0
        for i in range(len(chromosome)):
            if self.optimal_chromosome[i] == chromosome[i]:
                fitness += 1
        return fitness
    
    def swap_bits_population(self, new_population):
        for i in range(0, len(new_population)-1):
            num_bits = randint(1, min(len(new_population[i]), len(new_population[i+1]))-1)
            swap_indices = sample(range(len(new_population[i])), num_bits)
            for j in swap_indices:
                new_population[i] = new_population[i][:j] + new_population[i+1][j] + new_population[i][j+1:]
                new_population[i+1] = new_population[i+1][:j] + new_population[i][j] + new_population[i+1][j+1:]
        return new_population
  
    def generate_population(self):
        print("INITIAL POPULATION")
        for i in range(self.size_of_population):
            random_sample = f'{int(randint(3, 15)):08b}'
            if random_sample not in self.population:
                self.population.append(random_sample)
        print(self.population)
    
    def execute(self):
        self.iterations += 1
        print(f'\nEXECUTE: Number of iterations: {self.iterations}')
        if self.optimal_chromosome in self.population:
            print("EVOLVED")
            print("Optimal Chromosome: ", self.optimal_chromosome)
            x = self.population.index(self.optimal_chromosome)
            print("Evolved chromosome: ", self.population[x])
            self.population.append(self.population[x])
            del self.population[x]
            return self.population
        else:
            print("********>>> NO SOLUTION YET <<<********")
            self.cross_over()
            self.mutation()
            return self.execute()
    
    def cross_over(self):
        print("CROSSOVER", end=' ')
        temp = list()
        for chromosome in self.population:
            temp.append([self.fitness_check(chromosome), chromosome])
        temp.sort(key=lambda x:x[0])
        self.population.clear()
        for i in range(len(temp)):
            self.population.append(temp[-1][1])
            del temp[-1]
        self.population = self.swap_bits_population(self.population)
        print(self.population)
    
    def mutation(self):
        print("MUTATION")
        for i in range(1):
            index = randint(0, self.population.__len__()-1)
            chromosome = self.population[index]
            gene = randint(0, 1)
            chromosome = chromosome[5:]
            chromosome = chromosome + str(gene)
            chromosome = f'{int(chromosome, 2):08b}'
            print("Mutated Chromosome: ", chromosome)
            self.population[index] = chromosome
```
